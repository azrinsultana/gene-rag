```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
import re


abstracts = uniprot_df['clean_function'].dropna()

text = ' '.join(abstracts)


text = text.lower()
text = re.sub(r'[^a-z\s]', '', text)

extra_stopwords = {'also', 'however', 'although', 'thus', 'therefore', 'using', 'used', 'one', 'two', 'three', 'may', 'will', 'can', 'could', 'would', 'should', 'et', 'al'}
stopwords = set(STOPWORDS).union(extra_stopwords)

wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white',
    stopwords=stopwords,
    max_words=200,
    colormap='viridis',
    random_state=42
).generate(text)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.savefig('uniprot_wordcloud.png', dpi=300, bbox_inches='tight')

plt.show()

```
